### 1. 상자, 키 매칭시키기
* 김종성씨와 조영욱씨는 각각 상자와 키를 가지고 있습니다 상자와 키를 각각 매칭시켜 열수있는 함수를 만드세요
* 상자와 키의 색은 같아야 열립니다

```python
kim = ["redkey","bluekey","greenkey","yellowkey","blackkey"]
cho = ["blackbox","yellowbox","redbox","whitebox","graybox"]
```

```python
kim = ["redkey","bluekey","greenkey","yellowkey","blackkey"]
cho = ["blackbox","yellowbox","redbox","whitebox","greybox"]
kimvalue= set()
chovalue= set()
def boxkey(kim,cho):
    for i in kim:
        kimvalue.add(i[:-3])
    for i in cho:
        chovalue.add(i[:-3])
    return kimvalue & chovalue

boxkey(kim,cho)
```

### 2. 포커카드 숫자로 만들기
* 포커카드 J,Q,K,A 를 받을때마다 카운팅 하여 숫자로 반환하는 함수를 만들고 딕셔너리로 만드세요

```python
poker = ['A','J','Q','J','Q','Q','Q','J','A','K','A','A','A','Q','J']
```

```python
poker = ['A','J','Q','J','Q','Q','Q','J','A','K','A','A','A','Q','J']


def card(card):
    J = card.count("J")
    Q = card.count("Q")
    K = card.count("K")
    A = card.count("A")
    return {'J':J,'Q':Q,'K':K,'A':A}

print(card(poker))
```

* 참조 (randrange)

```python
from random import *

blackJack=0;
def blackjack():
    aCard = randrange(1,11)
    bCard = randrange(1,11)
    value = aCard+bCard
    if(value == 22):
        return "버스트하셨습니다"
    else:   
        if value <= 18:
            while value <= 18:
                value += randrange(1,11)
                if value >= 22:
                    return "버스트하셨습니다"
            return value   
        else:
            return value
        

blackjack()
```

### 5. 합의 제곱 빼기 제곱의 합
* 1부터 주어진 숫자까지의 합의 제곱에서 제곱의 합을 빼주세요

```python
num1 = 4
num2 = 30
num3 = 77
```

```python
def minus(num):
    value1 = 0
    value2 = 0
    for x in range(1, num + 1):
        value1 += x*x
    for y in range(1, num + 1):
        value2 += y
    return value2*value2 - value1

minus(num1)
```

```python
minus(num2)
```

```python
minus(num3)
```

### 6. 봄의 시작을 맞춰라!
* 주어진 list를 보고 봄이 시작하는 날짜를 맞춰주세요
* 1일 평균기온의 9일 이동 평균값이 5℃이상으로 올라간 첫 번째 날을 봄의 시작으로 봅니다
* 9일 이동 평균값은 과거 9일 동안 일평균기온의 평균입니다
* tempList는 한달간의 일 평균기온을 나타냅니다

```python
tempList1 = [3, 5, 6, 2, 5, 7, 8, 2 ,3 ,1, 0, 3, 4, 5, 4, 3, 7, 8, 4, 3, 5, 7, 8, 3, 9 ,6 ,7 ,5 ,8, 4]
tempList2 = [1, 2, 5, 1, 0, 2, 3, 4, 4, 8, 4, 1, 0, 4, 3, 4, 5, 3, 5, 7, 4, 6, 4, 2, 3, 4, 2, 3, 5, 3]
```

```python
def defineSpring(temperatureList):
    average = 0
    count = 0
    secondCount = 0
        
    while count < 9:
        average += temperatureList[count]
        count += 1
    if average // 9 >= 5:
        return str(count + 1) + "일에 봄이 시작되었습니다"

    while count < 30:
        average -= temperatureList[secondCount]
        average += temperatureList[count]
        count += 1
        secondCount += 1
        if average // 9 >= 5:
            return str(count + 1) + "일에 봄이 시작되었습니다"
    return "봄이 시작되지 않았습니다"

defineSpring(tempList1)
```

```python
defineSpring(tempList2)
```
